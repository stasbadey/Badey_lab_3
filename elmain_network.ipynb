```python
import numpy as np


class Elman:
    def __init__(self, sequence: list, p: int, error: float, max_iter: int, m: int, alpha: float, predict: int,
                 code_for_learning: str):
        self.sequence = sequence
        self.p = p
        self.error = error
        self.max_iter = max_iter
        self.m = m
        self.alpha = alpha
        self.predict = predict
        self.code_for_learning = code_for_learning

    @staticmethod
    def derivative_function(x):
        return 1

    @staticmethod
    def activation_function(x):
        return 1 * x

    def start(self):
        q = len(self.sequence)
        x = []
        y = []
        i = 0
        while i + self.p < q:
            x.append(self.sequence[i: i + self.p])
            y.append(self.sequence[i + self.p])
            i += 1
        y = np.array(y)
        x = np.array(x)
        error_all = 0
        k = 0
        if self.code_for_learning[0] == "1":
            context = np.zeros((x.shape[0], self.m))
        else:
            context = np.random.rand(x.shape[0], self.m)
        x = np.concatenate((x, context), axis=1)
        x = x.reshape(x.shape[0], 1, x.shape[1])
        w1 = (np.random.rand(self.p + self.m, self.m) * 2 - 1) / 10
        w2 = (np.random.rand(self.m, 1) * 2 - 1) / 10

        self.learn_weights(self.max_iter, self.code_for_learning, w1, w2, self.alpha, x, y, self.m, self.error)

        print(error_all)
        k = y[-1].reshape(1)
        X = x[-1, 0, :-self.m]
        out = []
        X = X[1:]
        train = np.concatenate((X, k))
        X = np.concatenate((X, k))
        train = np.append(train, np.array([0] * self.m))
        hidden_layer = np.matmul(train, w1)
        output = np.matmul(hidden_layer, w2)
        k = output
        out.append(k[0])
        # self.save_result(self.sequence, out)
        return out

    def learn_weights(self, max_iter, code_for_learning, w1, w2, alpha, x, y, m, error):
        for j in range(max_iter):
            error_all = 0
            if code_for_learning[1] == "1":
                x[:, :, -m:] = 0
            for i in range(x.shape[0]):
                hidden_layer = self.activation_function(np.matmul(x[i], w1))
                output = self.activation_function(np.matmul(hidden_layer, w2))
                dy = output - y[i]
                w1 -= alpha * dy * np.matmul(x[i].transpose(), w2.transpose()) * self.derivative_function(
                    np.matmul(x[i], w1))
                w2 -= alpha * dy * hidden_layer.transpose() * self.derivative_function(np.matmul(hidden_layer, w2))
                try:
                    x[i + 1][-m:] = hidden_layer
                except:
                    pass
            for i in range(x.shape[0]):
                hidden_layer = np.matmul(x[i], w1)
                output = np.matmul(hidden_layer, w2)
                dy = output - y[i]
                error_all += (dy ** 2)[0]
            print(j + 1, " ", error_all[0])
            if error_all <= error:
                break
```
