```python
import numpy as np


def matrix_product(vector_1, vector_2):
    return np.matmul(vector_1, vector_2)


def from_list_to_array(x):
    return np.array(x)
```

```python
class Elman:
    def __init__(self, sequence: list, p: int, error: float, max_iter: int, m: int, alpha: float, predict: int):
        self.sequence = sequence
        self.p = p
        self.error = error
        self.max_iter = max_iter
        self.m = m
        self.alpha = alpha
        self.predict = predict
        self.x = []
        self.y = []
        self.q = len(self.sequence)
        self.w1 = None
        self.w2 = None

    @staticmethod
    def derivative_function(x):
        return 1

    @staticmethod
    def activation_function(x):
        return 1 * x

    def append_sequence(self):
        i = 0
        while i + self.p < self.q:
            self.x.append(self.sequence[i: i + self.p])
            self.y.append(self.sequence[i + self.p])
            i += 1

    def start(self):
        self.append_sequence()
        y = from_list_to_array(self.y)
        x = from_list_to_array(self.x)
        context = np.random.rand(x.shape[0], self.m)
        x = np.concatenate((x, context), axis=1)
        x = x.reshape(x.shape[0], 1, x.shape[1])
        self.w1 = np.random.rand(5, 2)
        self.w2 = np.random.rand(2, 1)

        self.learn_weights(x, y)

        print(self.result_processing(x, y))

    def result_processing(self, x, y):
        k = y[-1].reshape(1)
        X = x[-1, 0, :-self.m]
        out = []
        X = X[1:]
        train = np.concatenate((X, k))
        train = np.append(train, np.array([0] * self.m))
        hidden_layer = matrix_product(train, self.w1)
        output = matrix_product(hidden_layer, self.w2)
        out.append(output[0])
        return out

    def weight_processing(self, dy, x, i, hidden_layer):
        self.w1 -= self.alpha * dy * matrix_product(x[i].transpose(),
                                                    self.w2.transpose()) * self.derivative_function(
            matrix_product(x[i], self.w1))
        self.w2 -= self.alpha * dy * hidden_layer.transpose() * self.derivative_function(
            matrix_product(hidden_layer, self.w2))

    def learn_weights(self, x, y):
        j = 0
        while j < self.max_iter:
            error_all = 0
            i = 2
            x[:, :, -self.m:] = 0
            hidden_layer = self.activation_function(matrix_product(x[i], self.w1))
            output = self.activation_function(matrix_product(hidden_layer, self.w2))
            dy = output - y[i]
            self.weight_processing(dy, x, i, hidden_layer)
            hidden_layer = matrix_product(x[i], self.w1)
            output = matrix_product(hidden_layer, self.w2)
            dy = output - y[i]
            error_all += (dy ** 2)[0]
            j += 1
            print(j, " ", error_all[0])
            if error_all <= self.error:
                print("last error = ", error_all)
                break
```
