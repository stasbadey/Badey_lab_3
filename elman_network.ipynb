```python
import numpy as np


class Elman:
    def __init__(self, sequence: list, p: int, error: float, max_iter: int, m: int, alpha: float, predict: int):
        self.sequence = sequence
        self.p = p
        self.error = error
        self.max_iter = max_iter
        self.m = m
        self.alpha = alpha
        self.predict = predict
        self.x = []
        self.y = []
        self.q = len(self.sequence)
        self.w1 = None
        self.w2 = None

    @staticmethod
    def derivative_function(x):
        return 1

    @staticmethod
    def activation_function(x):
        return 1 * x

    def start(self):
        i = 0
        while i + self.p < self.q:
            self.x.append(self.sequence[i: i + self.p])
            self.y.append(self.sequence[i + self.p])
            i += 1
        y = np.array(self.y)
        x = np.array(self.x)
        context = np.random.rand(x.shape[0], self.m)
        x = np.concatenate((x, context), axis=1)
        x = x.reshape(x.shape[0], 1, x.shape[1])
        self.w1 = np.random.rand(5, 2)
        self.w2 = np.random.rand(2, 1)

        self.learn_weights(x, y)

        k = y[-1].reshape(1)
        X = x[-1, 0, :-self.m]
        out = []
        X = X[1:]
        train = np.concatenate((X, k))
        train = np.append(train, np.array([0] * self.m))
        hidden_layer = np.matmul(train, self.w1)
        output = np.matmul(hidden_layer, self.w2)
        k = output
        out.append(k[0])
        return out

    def learn_weights(self, x, y):
        for j in range(self.max_iter):
            error_all = 0
            x[:, :, -self.m:] = 0
            for i in range(x.shape[0]):
                hidden_layer = self.activation_function(np.matmul(x[i], self.w1))
                output = self.activation_function(np.matmul(hidden_layer, self.w2))
                dy = output - y[i]
                self.w1 -= self.alpha * dy * np.matmul(x[i].transpose(), self.w2.transpose()) * self.derivative_function(
                    np.matmul(x[i], self.w1))
                self.w2 -= self.alpha * dy * hidden_layer.transpose() * self.derivative_function(np.matmul(hidden_layer, self.w2))
                hidden_layer = np.matmul(x[i], self.w1)
                output = np.matmul(hidden_layer, self.w2)
                dy = output - y[i]
                error_all += (dy ** 2)[0]
            print(j + 1, " ", error_all[0])
            if error_all <= self.error:
                print("last error = ", error_all)
                break

```
